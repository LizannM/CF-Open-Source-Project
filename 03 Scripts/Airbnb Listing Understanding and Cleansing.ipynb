# 8. Histograms: price and minimum_nights

```python
# prices histogram
hist_price = listings['price'].plot.hist(bins=100)
```

There are outliers with prices into the thousands of dollars, which could be legitmate listings or a data error

```python
# Check for prices above $7,000
listings.loc[listings['price'] > 7000]
```

```python
# Mark prices over $7000 as NaN
listings.loc[listings['price'] > 7000, 'price'] = np.nan
```

```python
listings['price'].max()
```

```python
hist_price2 = listings['price'].plot.hist(bins=350)
```

```python
hist_min = listings['minimum_nights'].plot.hist(bins=25)
```

```python
# Mark nights over 90 as NaN
listings.loc[listings['minimum_nights'] > 90, 'minimum_nights'] = np.nan
```

```python
hist_min2 = listings['minimum_nights'].plot.hist(bins=25)
```
